### Objetivos:

#### Objetivo 1: Agente Reactivo (Arquitectura Simple)

Diseñe la lógica para el Agente Reactivo. Este agente solo utiliza reglas de umbral basadas en la percepción inmediata (no planifica a largo plazo ni considera la utilidad del clima).

* Regla de Batería (Prioridad Alta): Si la batería es inferior a 15%, la única acción posible es Cargar_Bateria.
* Regla de Movimiento (Prioridad Baja): Si la batería es suficiente, el agente solo debe moverse en la dirección que reduzca la distancia a su destino (la dirección "codiciosa" o greedy).
  
  Pregunta Clave: ¿Qué sucede si hay niebla o si el destino está lejos y la batería está a $16\%$? ¿Es este agente eficiente?

#### Objetivo 2: Agente Basado en Utilidad (Arquitectura Racional)

Diseñe la lógica para el Agente Basado en Utilidad. Este agente debe elegir la acción que maximice la Utilidad Esperada, considerando tres factores ponderados:


1. Utilidad de Entrega (U_entrega): Maximizar el progreso hacia el destino.

2. Utilidad de Seguridad (U_seguridad): Penalizar las acciones arriesgadas (baja batería o mal clima).

3. Utilidad de Tiempo (U_tiempo): Penalizar las acciones lentas (Esperar o Cargar_Bateria).


#### Criterios de Utilidad Sugeridos:

* Ponderaciones: Usar $\text{W\_Entrega} = 0.6$, $\text{W\_Seguridad} = 0.3$, $\text{W\_Tiempo} = 0.1$.
* Movimiento: Las acciones de movimiento aumentan $U_{entrega}$ si acercan al destino (por ejemplo, $+0.8$).
* Seguridad:
  - Si el clima es 'Niebla' o 'Lluvia', todas las acciones de movimiento tienen una penalización de $-0.5$ en la seguridad.
  - Si la batería está entre $15\%$ y $30\%$, todas las acciones de movimiento tienen una penalización de $-0.3$ en la seguridad.
* Tiempo: La acción Cargar_Bateria tiene una penalización de $-0.9$ en la utilidad de tiempo.

Pregunta Clave: ¿Cómo se comporta este agente con un $20\%$ de batería y 'Niebla'? ¿Prioriza moverse, cargar o esperar? ¿Por qué?
